Loads the cleaned roads GeoJSON and renders a quick plot for sanity-checking coverage.

```python
import geopandas as gpd
import matplotlib.pyplot as plt

roads = gpd.read_file("../data/osm_roads.geojson")
ax = roads.plot(figsize=(8, 8), linewidth=0.2, color="firebrick")
ax.set_title("OSM Roads within Study Area")
ax.set_axis_off()
plt.show()
```
